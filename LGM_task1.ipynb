```python
df['Species'] = df['Species'].astype('string')
df.dtypes
```

```python
df['SepalL'] = df['SepalL'].round(1)
df['SepalW'] = df['SepalW'].round(2)
df['PetalL'] = df['PetalL'].round(2)
df['PetalW'] = df['PetalW'].round(2)
df.head()
```

```python
ndf = df.fillna(method  = 'ffill')
```

```python
plt.hist(ndf['SepalL'])
```

```python
ndf
```

```python
Dummies = pd.get_dummies(ndf['Species'])
Dummies
```

```python
sorted = pd.concat([ndf,Dummies], axis= 1)
sorted
```

```python
sorted.drop('Species', axis = 1, inplace = True)
sorted
```

```python
X  = sorted[['SepalL', 'PetalL', 'SepalW','PetalW']]
y = sorted[['Iris-setosa']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
R = LinearRegression()
R.fit(X_train, y_train)
```

```python
print(R.score(X_test, y_test))
```

```python
pred = R.predict(X_test)
pred
```
